# Exploratory Data Analysis (EDA)

This notebook performs EDA on a dataset with the following columns:

- id
- brand
- model
- model_year
- mileage
- fuel_type
- engine
- transmission
- ext_col
- int_col
- accident
- clean_title
- price

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset (Assuming the dataset is named 'car_data.csv')
data = pd.read_csv("./../data/raw/train.csv")

# Display the first few rows of the dataset
data.head()
```

```python
# Display basic information about the dataset
data.info()

# Display summary statistics
data.describe(include="all")
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])

# Optionally: Handle missing values (e.g., drop or fill)
data = data.dropna()  # This line can be modified as needed
```

```python
# Visualize the distribution of categorical columns
categorical_cols = [
    "brand",
    "model",
    "fuel_type",
    "transmission",
    "ext_col",
    "int_col",
    "accident",
    "clean_title",
]

for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=data, x=col, palette="viridis")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Visualize the distribution of numerical columns
numerical_cols = ["model_year", "milage", "engine", "price"]

for col in numerical_cols:
    plt.figure(figsize=(10, 4))
    sns.histplot(data=data, x=col, kde=True, bins=30, color="blue")
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Select only numerical columns for the correlation matrix
numerical_data = data.select_dtypes(include=["number"])

# Correlation matrix
corr_matrix = numerical_data.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
